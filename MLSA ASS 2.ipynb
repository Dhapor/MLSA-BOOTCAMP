##### 2. Python Problem: A Simple Inventory System
Problem:
Create a Python program to manage a simple inventory of products. The program should allow users to:

Add new products to the inventory
Remove existing products
Update product quantities
Search for products by name
Display the entire inventory

Expand the previous inventory system to include the following features:

Product Categories: Allow users to assign products to categories (e.g., "Electronics," "Clothing," "Food").
Product Prices: Store the price of each product.
Quantity Thresholds: Set a minimum quantity threshold for each product, and alert the user when the quantity falls below the threshold.
Sales: Track sales of products, including the quantity sold and total revenueÂ generated.

```python
class Inventory_system:
    def __init__(self):
        self.products = []

    def add(self):
        name = input('Enter the product name ( e.g., "Apple"): ').strip()
        price = float(input("Enter the product price (e.g., '19.00'): ").strip())
        category = input('Enter the product caterogy (e.g "electronics", "clothings", "food", etc.): ').strip()
        while True:
            quantity = int(input("Enter the product quantity, it should be above 10 (e.g., '10 and above'): ").strip())

            if quantity < 10:
                self.products.append({"name": name, "quantity": quantity, "price": price, 'category': category})
                print(f"Product '{name}' has been added successfully")
                break
            else:
                print(f"{quantity} is too small, input quantity above the set threshold")
                


    def remove(self):
        name = input("Enter the product name you wish to update ( e.g., 'Apple'): ").strip()
        for i in self.products:
            if i["name"].lower() == name.lower():
                self.products.remove(i)
                print(f"Product {name} removed succesfully")
                return self.products
            

    def update(self):
        name = input("Enter the product name you wish to update ( e.g., 'Apple'): ").strip()
        for i in self.products:
            if i["name"].lower() == name.lower():
                new_quantity = int(input("Enter the new quantity (e.g., '50'): "))
                i["quantity"] = new_quantity
                print(f"Product {name} has been successfully updated to {new_quantity}")
                return self.products
    

    def search(self):
        name = input("Enter name of product you want to search for: ").strip
        for i in name:
            if i in self.products:
                if i["name"].lower() == name.lower():
                    print(f"Product found: Name: {i['name']}, Quantity: {i['quantity']}, Price: {i['price']},")
            else:
                print(f"Product name '{name}' no found")
    
    def view_inventory(self):
        print(self.products)




    def display_help(self):
        print("\nHow to Use the Inventory Management System:")
        print("1. Product names should be entered as text, e.g., 'Apple'.")
        print("2. Quantities should be entered as whole numbers, e.g., '50'.")
        print("3. Prices should be entered as decimal numbers, e.g., '0.99'.")
        print("4. Use the menu to navigate through different options.")

    def display_menu(self):
        print("\nInventory Management System")
        print("1. View Inventory")
        print("2. Add New Product")
        print("3. Remove Product")
        print("4. Update Product Quantity")
        print("5. Search for Product")
        print("6. Help")
        print("7. Exit")

    def run(self):
        while True:
            self.display_menu()
            choice = input("Enter your choice (1-7): ").strip()

            if choice == "1":
                self.view_inventory()
            elif choice == "2":
                self.add()
            elif choice == "3":
                self.remove()
            elif choice == "4":
                self.update()
            elif choice == "5":
                self.search()
            elif choice == "6":
                self.display_help()
            elif choice == "7":
                print("Exiting the program. Goodbye!")
                break
            else:
                print("Invalid choice. Please enter a number between 1 and 7.")


# Run the Inventory Management System
if __name__ == "__main__":
    IVS = Inventory_system()
    IVS.run()
```
